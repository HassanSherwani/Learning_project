Now let's cast zip code from a float to a string. Some zip codes can start with 0 (not in Chicago), and we need to account for that. Zip is integter some states start with zero. So, we need data with string.

```python
import numpy as np


def float_to_zip(zip_code):
    if np.isnan(zip_code):
        return np.nan
    
    # 0 makes sure to left-pad with zero
    # zip codes have 5 digits
    # .0 means, we don't want anything after the decimal
    # f is for float
    zip_code = "{:05.0f}".format(zip_code)
    return zip_code
```

```python
dta.zip = dta.zip.apply(float_to_zip)
```
