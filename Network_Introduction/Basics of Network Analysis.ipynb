# Basics of Network Analysis

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
```

```python
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)
G.add_node(6)
```

```python
G.nodes()
```

```python
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(4,6)
G.add_edge(5,4)
G.add_edge(2,3)
```

```python
G.edges
```

```python
G.add_edge(2,6)
```

```python
G.edges
```

```python
# Visualize graph
nx.draw(G)
plt.show
```

```python
nx.draw(G, with_labels=1)
plt.show
```

```python
z=nx.complete_graph(10) # graph of 10 vertices 
```

```python
z.nodes
```

```python
z.edges
```

```python
z.order() # no. of nodes in graph
```

```python
z.size() # no. of edges
```

```python
nx.draw(z, with_labels=1)
plt.show
```

```python
# Graph on 100 vertices
H=nx.complete_graph(100)
```

```python
nx.draw(H)
plt.show()
```

```python
# Generating 20 vertices with probability of 0.5. Here edges are drawn randomly
I=nx.gnp_random_graph(20,0.5)
```

```python
nx.draw(I)
plt.show()
```
