# Modeling Road network between cities

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

## Creating a graph

```python
G=nx.Graph() # undirected graph
#X=nx.DiGraph() #Directed Graph
```

## Choosing cities

```python
City_set=['Amsterdam','Paris','London','Madrid','Helsinki','Budapest','Vienna', 'Copenhagen']
```

```python
print(City_set)
```

```python
for each in City_set:
    G.add_node(each)
```

```python
# Visualize graph
nx.draw(G, with_labels=1)
plt.show
```

## Adding edges i.e rodes between these cities

```python
# Creating an array of edges
roads=[]
value=100
while (value <= 2000):
    roads.append(value)
    value=value+100
print (roads)
```

```python
# Let's add 16 edges to our city network
while(G.number_of_edges() < 16):
    c1 = random.choice(list(G.nodes()))
    c2 = random.choice(list(G.nodes()))
    if c1 != c2 and G.has_edge(c1, c2) == 0:    # there should be no same edge between city and cities should be different
        w = random.choice(roads)
        G.add_edge(c1, c2, weight=w)
```

```python
# Visualize graph
nx.draw(G, with_labels=1)
plt.show
```

```python
# Another way is through spectral
pos=nx.spectral_layout(G)
nx.draw(G, pos, with_labels= 1)
plt.show()
```

```python
# Spring layout
pos=nx.spring_layout(G)
nx.draw(G, pos, with_labels= 1)
plt.show()
```

```python
# Circular layout
pos=nx.circular_layout(G)
nx.draw(G, pos, with_labels= 1)
plt.show()
```

```python
# For edge weights
pos=nx.circular_layout(G)
nx.draw(G)
nx.draw_networkx_edge_labels(G, pos)
plt.show()
```

## Checking paths between cities

```python
# check if graph is connected or not
nx.is_connected(G)
```

```python
# Let's reduce number of edges
G2=nx.Graph() # undirected graph
```

```python
for each in City_set:
    G2.add_node(each)
```

```python
while(G2.number_of_edges() < 5):
    c1 = random.choice(list(G2.nodes()))
    c2 = random.choice(list(G2.nodes()))
    if c1 != c2 and G2.has_edge(c1, c2) == 0:    # there should be no same edge between city and cities should be different
        w = random.choice(roads)
        G2.add_edge(c1, c2, weight=w)
```

```python
nx.is_connected(G2)
```

```python
pos=nx.circular_layout(G2)
nx.draw(G2,pos,with_labels= 1)
plt.show()
```

```python
# Suppose we want to see which path is connected and which path is not
for u in G2.nodes():
    for v in G2.nodes():
        print (u, v, nx.has_path(G2,u,v))
```

## Shortest Path

```python
G3 = nx.Graph()
for each in City_set:
    G3.add_node(each)
        
while(G3.number_of_edges() < 10):
        c1 = random.choice(list(G3.nodes()))
        c2 = random.choice(list(G3.nodes()))
        if c1 != c2 and G3.has_edge(c1, c2) == 0:  
            w = random.choice(roads)
            G3.add_edge(c1, c2, weight=w)
```

```python
pos=nx.circular_layout(G3)
nx.draw(G3, with_labels= 1)
nx.draw_networkx_edge_labels(G3, pos)
plt.show()
```

```python
# Choose two nodes for finding shortest path. We can even check for first graph
u='Amsterdam'
v='Budapest'
print(nx.dijkstra_path(G3,u,v))
```

```python
# For path length
print(nx.dijkstra_path_length(G3,u,v))
```
